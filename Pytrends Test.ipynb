```python
pytrend = TrendReq()
```

```python
pytrend.build_payload(kw_list=['vlog','blog'],timeframe='today 12-m', geo ='ID')
```

```python
interest_over_time_df = pytrend.interest_over_time()
```

```python
print(interest_over_time_df.head())
```

```python
#simpen ke local
print(interest_over_time_df.to_csv('file2.csv'))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
```

```python
dx = interest_over_time_df.plot.line(figsize =(9,6), title="Interest Over Time Blog Vs Vlog")
dx.set_xlabel('Tanggal')
dx.set_ylabel('Index Tren')
```
