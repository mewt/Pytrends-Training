```python
pytrend = TrendReq()
```

```python
pytrend.build_payload(kw_list=['vlog','blog'], timeframe='today 5-y', geo = 'ID')
```

```python
interest_over_time_df = pytrend.interest_over_time()
```

```python
print(interest_over_time_df.head())
```

# Print to CSV

```python
print(interest_over_time_df.to_csv('file3.csv'))
```
